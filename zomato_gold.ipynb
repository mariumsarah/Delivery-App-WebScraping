```python
from bs4 import BeautifulSoup
import requests
import re
import xlsxwriter 
# Because Zomato Access denied
def extract_source(url):
     agent = {"User-Agent":"Mozilla/5.0"}
     source=requests.get(url, headers=agent).text
     return source
```

```python
workbook = xlsxwriter.Workbook('zomato_gold.xlsx') 
# The workbook object is then used to add new  
# worksheet via the add_worksheet() method. 
worksheet = workbook.add_worksheet() 
worksheet.write(0,0,'RestaurantID')
worksheet.write(0,1,'Chain')
worksheet.write(0,2,'Type')
worksheet.write(0,3,'Category')
worksheet.write(0,4,'City')
worksheet.write(0,5,'Feature')
worksheet.write(0,6,'Cuisine')
worksheet.write(0,7,'Area')
worksheet.write(0,8,'Price for Two')
worksheet.write(0,9,'Address')
worksheet.write(0,10,'Rating')
worksheet.write(0,11,'Review')
worksheet.write(0,12,'Timing')
worksheet.write(0,13,'Rating 2')
worksheet.write(0,14,'Review 2')
worksheet.write(0,15,'Offers & Discounts')
#worksheet.write(0,16,'Reviewed By')
```

```python
# CHANGE CITY NAME ACCORDING TO WHAT IS NEEDED
city_names = ['ajman','al-ain','ras-al-khaimah','fujairah','umm-al-quwain','Dubai','doha','Sharjah','abudhabi']
# NUMBER OF PAGES TO GO THROUGH STARTING FROM PAGE 1 
pagess = [82,74,59,26,17,477,163,150,270]
first = 0
# For City
for city in range(len(city_name)): 
    city_name = city_names[city]
    pages = pagess[city]
    
    # For Page
    for x in range(1,pages):

        print(city_name+' ----------- On Page '+str(x)+' --------------------------------')
        soup = BeautifulSoup(extract_source("https://www.zomato.com/"+city_name+"/restaurants?page="+str(x)), 'html.parser')
        second = 0
        for x in soup.find_all('div', class_='content'):
            if second >= 1:
                
                # ----------------------------- Chain Name
                try:
                    x.find_all('a', class_='result-title')[0].get_text().rstrip()
                    worksheet.write(first,1,x.find_all('a', class_='result-title')[0].get_text().rstrip() )
                except IndexError:
                    first = first - 1
                    break
                
                print('Restaurant '+str(first))
                # ----------------------------- RestaurantID
                worksheet.write(first,0,str(first))

                # ----------------------------- Type: Sponsored / Not Sponsored
                try:
                    worksheet.write(first,2,x.find_all('div',class_='zgold')[0].text)
                except IndexError: 
                    worksheet.write(first,2,'NOT SPONSORED')

                # ----------------------------- Casual Dining/ Quick Bite
                try:
                    worksheet.write(first,3,x.find_all('div', class_='res-snippet-small-establishment')[0].get_text().rstrip())
                except IndexError:
                    worksheet.write(first,3,'Not Available')

                # ----------------------------- City Name
                worksheet.write(first,4,city_name)


                # ----------------------------- FEATURED IN
                try:
                    worksheet.write(first,5,x.find_all('div',class_='res-collections clearfix')[0].get_text().rstrip().replace("\n","").replace('Featured in: ',''))
                except IndexError:
                    worksheet.write(first,5,'Not Available')

                # ----------------------------- CUISINES
                worksheet.write(first,6, x.find_all('span',class_='col-s-11 col-m-12 nowrap pl0')[0].get_text().rstrip())

                # ----------------------------- ZONES AVAILABE IN
                worksheet.write(first,7,x.find_all('a', class_='search_result_subzone')[0].get_text().rstrip())

                # ----------------------------- COST FOR TWO
                try:
                    worksheet.write(first,8,x.find_all('div',class_='res-cost clearfix')[0].text.replace('Cost for two:',''))
                except IndexError: 
                    worksheet.write(first,8,'Not Available')

                # ----------------------------- ADDRESS
                worksheet.write(first,9,x.find_all('div', class_='search-result-address')[0].get_text().rstrip())

                # ----------------------------- RATING
                try: 
                    worksheet.write(first,10,(x.find_all('span', class_='rating-value')[0].get_text().rstrip()))
                except IndexError: 
                    worksheet.write(first,10,'Not Available')
                # ----------------------------- REVIEW
                try: 
                    worksheet.write(first,11,(x.find_all('span', class_='review-count')[0].get_text().rstrip()).replace('(','').replace(')',''))
                except IndexError: 
                    worksheet.write(first,11,'Not Available')

                # ----------------------------- RATING 2 
                try: 
                    worksheet.write(first,13,(x.find_all('span', class_='rating-value')[1].get_text().rstrip()))
                except IndexError: 
                    worksheet.write(first,13,'Not Available')
                # ----------------------------- REVIEW 2 
                try: 
                    worksheet.write(first,14,(x.find_all('span', class_='review-count')[1].get_text().rstrip().replace('(','').replace(')','')))
                except IndexError: 
                    worksheet.write(first,14,'Not Available')

                # ----------------------------- TIMING
                try: 
                    worksheet.write(first,12,x.find_all('div',class_='res-timing clearfix')[0].get("title"))
                except IndexError: 
                    worksheet.write(first,12,'Not Available')
                    
                # ----------------------------- Offers & discounts
                try: 
                    worksheet.write(first,15,', '.join([r.get_text().rstrip().replace('\n','').replace('  ','') for r in x.find_all('div',class_='res-offers clearfix')]))
                except IndexError: 
                    worksheet.write(first,15,'Not Available')
                    
                # ----------------------------- RECOMMENDED  BY 
#                 try: 
#                     #worksheet.write(first,15,
#                     print(x.find_all('div',class_='res-recommended clearfix')[0])
#                 except IndexError: 
#                     print('nvm')
                    #worksheet.write(first,15,'Not Available')                    
            first+=1
            second += 1
```
